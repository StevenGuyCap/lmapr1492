## Mise en place du code

```python
# T√©l√©chargement du mat√©riau attribu√©

from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter

k = "zkUgCxiKVvQri3sF2Sx"  # Cl√© de l'API

with MPRester(k) as m:
    structure = m.get_structure_by_material_id("mp-31454")  # R√©cup√©ration de la structure assign√©e
```

```python
# Import(s) n√©cessaire(s) pour la suite du notebook

from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
# Instanciation du calculateur de diffractions des rayons X de longeur d'onde type CuKalpha (lambda = 1.54060 √Ö)
xrd = XRDCalculator('CuKa')

# R√©cup√©ration des r√©sultats de la diffraction
pattern = xrd.get_pattern(structure)
```

---

## Analyse

<br />

En sachant que la variable *pattern* correspond √† un objet de la classe *DiffractionPattern*, les diff√©rents champs contenus dans celui-ci nous donne les informations n√©cessaires afin d'analyser notre diffractogramme, ou du moins, dans le cas de ce notebook, les informations n√©cessaires pour r√©pr√©senter les 3 premiers pics de celui-ci.

Cette classe contient les champs :
* *x* : correspond √† une liste des angles $2\theta$ li√©s aux pics de la diffraction
* *y* : correspond √† une liste des valeurs des intensit√©s des pics de la diffraction
* *hkls* : correspond √† une liste contenant des objets dictionnaires avec les champs "hkl" correspond aux plans de diffraction et "mult" correspondant √† la multiplicit√© de ces pics
* *d_hkls* : correspond √† une liste des distances inter-plan

---
### Premier pic

```python
# Position selon l'angle 2 theta du premier pic
xp1 = pattern.x[0]

# Intensit√© du premier pic
yp1 = pattern.y[0]

# Plans de diffraction du premier pic
hklsp1 = pattern.hkls[0]

# Distance inter-plan pour le premier pic
dhklsp1 = pattern.d_hkls[0]

print(75*"=")
print("R√©sultats obtenus pour le premier pic du diffractogramme : ")
print(60*"-")
print("Angle 2ùúÉ             = {0:0.5f}¬∞".format(xp1))
print("Intensit√© du pic     = {0:0.5f}".format(yp1))
print("Plans de diffraction = {0}".format(hklsp1))
print("Distance inter-plan  = {0:0.5f} √Ö".format(dhklsp1))
print(75*"=")
```

---
### Second pic

```python
# Position selon l'angle 2 theta du second pic
xp2 = pattern.x[1]

# Intensit√© du second pic
yp2 = pattern.y[1]

# Plan de diffraction du second pic
hklsp2 = pattern.hkls[1]

# Distance inter-plan pour le second pic
dhklsp2 = pattern.d_hkls[1]

print(75*"=")
print("R√©sultats obtenus pour le second pic du diffractogramme : ")
print(60*"-")
print("Angle 2ùúÉ             = {0:0.5f}¬∞".format(xp2))
print("Intensit√© du pic     = {0:0.5f}".format(yp2))
print("Plans de diffraction = {0}".format(hklsp2))
print("Distance inter-plan  = {0:0.5f} √Ö".format(dhklsp2))
print(75*"=")
```

---
### Troisi√®me pic

```python
# Position selon l'angle 2 theta du troisi√®me pic
xp3 = pattern.x[2]

# Intensit√© du troisi√®me pic
yp3 = pattern.y[2]

# Plan de diffraction du troisi√®me pic
hklsp3 = pattern.hkls[2]

# Distance inter-plan pour le troisi√®me pic
dhklsp3 = pattern.d_hkls[2]

print(75*"=")
print("R√©sultats obtenus pour le troisi√®me pic du diffractogramme : ")
print(60*"-")
print("Angle 2ùúÉ             = {0:0.5f}¬∞".format(xp3))
print("Intensit√© du pic     = {0:0.5f}".format(yp3))
print("Plans de diffraction = {0}".format(hklsp3))
print("Distance inter-plan  = {0:0.5f} √Ö".format(dhklsp3))
print(75*"=")
```
